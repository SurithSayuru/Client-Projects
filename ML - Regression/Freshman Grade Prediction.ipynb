```python
x = df.drop(columns=['LAST NAME','freshman_grade'])
# x = df[['SHS_grade']]
y = df['freshman_grade']
```

```python
x
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=50)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create the MinMaxScaler
scaler = MinMaxScaler()

# Fit on training data and transform both training and test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import pickle

# Save the MinMaxScaler object to a file
with open('/content/drive/MyDrive/Colab Notebooks/Fiverr/arnelalexander/minmax_scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

Linear Regression model

```python
# Create the linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train_scaled, y_train)

# Make predictions on the training and test data
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# Evaluate the model on training data
train_metrics = {
    'Mean Absolute Error': mean_absolute_error(y_train, y_train_pred),
    'Mean Squared Error': mean_squared_error(y_train, y_train_pred),
    'Root Mean Squared Error': np.sqrt(mean_squared_error(y_train, y_train_pred)),
    'R-squared': r2_score(y_train, y_train_pred)
}

# Evaluate the model on test data
test_metrics = {
    'Mean Absolute Error': mean_absolute_error(y_test, y_test_pred),
    'Mean Squared Error': mean_squared_error(y_test, y_test_pred),
    'Root Mean Squared Error': np.sqrt(mean_squared_error(y_test, y_test_pred)),
    'R-squared': r2_score(y_test, y_test_pred)
}

# Display metrics as a table
metrics_df = pd.DataFrame({
    'Training Set': train_metrics,
    'Testing Set': test_metrics
})


metrics_df
```

```python
# Save the MinMaxScaler object to a file
with open('/content/drive/MyDrive/Colab Notebooks/Fiverr/arnelalexander/model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
# compare predictions and actuals in a table
comparision = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})
comparision
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming your DataFrame is named df
actual = comparision['Actual']
predicted = comparision['Predicted']

# Create an array for the x-axis positions
x = np.arange(len(comparision))

# Set the width of the bars
width = 0.35

plt.figure(figsize=(10, 5))

# Plot the actual grades
plt.bar(x - width/2, actual, width, label='Actual')

# Plot the predicted grades next to the actual grades
plt.bar(x + width/2, predicted, width, label='Predicted')

# Customize the chart
plt.xlabel('Instances')
plt.ylabel('Grades')
plt.title('Actual vs Predicted Grades')
plt.legend()

# Show the plot
plt.show()
```
