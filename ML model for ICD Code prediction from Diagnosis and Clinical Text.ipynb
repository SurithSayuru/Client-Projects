```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
import pandas as pd

# Assuming your data is in a DataFrame named df
# If your data is in a CSV file, you can read it using pd.read_csv('your_file.csv')

# Set the random seed for reproducibility
seed_value = 42
import random
random.seed(seed_value)

# Grouping by ICD9_CODE and selecting 50 random rows for each group (or all rows if less than 50)
def random_sample(group):
    return group.sample(min(len(group), 25), random_state=seed_value)

# Applying the random_sample function to each group
random_samples = df.groupby('ICD9_CODE', group_keys=False).apply(random_sample)

# Counting the occurrences of each ICD9_CODE after sampling
icd9_counts_after_sampling = random_samples['ICD9_CODE'].value_counts()

# Filtering out ICD9_CODE groups with less than 50 rows
icd9_codes_gt_50 = icd9_counts_after_sampling[icd9_counts_after_sampling >= 25].index

# Filtering the random_samples DataFrame to include only ICD9_CODE groups with at least 50 rows
final_result = random_samples[random_samples['ICD9_CODE'].isin(icd9_codes_gt_50)]

# Displaying the resulting DataFrame
final_result
```

```python
final_result.nunique()
```

```python
df_s = final_result
```

```python
# Specify your target variable
target_variable = 'ICD9_CODE'

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_s['Combined_Text'], df_s[target_variable], test_size=0.2, random_state=42)
```

```python
# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# Save the fitted TfidfVectorizer
with open('/content/drive/MyDrive/Colab Notebooks/Fiverr/Mohammad Anas/tfidf_vectorizer.pkl', 'wb') as vectorizer_file:
    pickle.dump(tfidf_vectorizer, vectorizer_file)

# Transform the test data
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Create and train the Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_tfidf, y_train)
```

```python
# Make predictions on the test set
y_pred_train = nb_classifier.predict(X_train_tfidf)

# Evaluate the model
accuracy_train = accuracy_score(y_train, y_pred_train)

print(f"Train Accuracy: {accuracy_train}")
```

```python
# Make predictions on the test set
y_pred = nb_classifier.predict(X_test_tfidf)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)

print(f"Test Accuracy: {accuracy}")
```

```python
import re
import string

def preprocess_text(text):
    # Remove date and time patterns
    text = re.sub(r'\[\*\*.*?\*\*\]', '', text)

    # Remove special characters and punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Convert to lowercase
    text = text.lower()

    # Tokenization (split the text into words)
    tokens = text.split()

    # Join the tokens back into a string
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

preprocessed_example = preprocess_text(clinical_text_value)
print(preprocessed_example)
```

```python
df['Clinical_Text'] = df['Clinical_Text'].apply(preprocess_text)
```
